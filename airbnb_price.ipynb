# извлечение данных и графический анализ

```python
data_t = data[{'Neighborhood Group', 'Property Type', 'Room Type', 'Accomodates', 'Bathrooms', 'Bedrooms', 'Beds', 'Price', 'Guests Included', 'Reviews', 'Overall Rating'}]
data_t = data_t.dropna()
```

исправить значения цены от str до float, например (5.000,00 -> 5000,00)
удалить строки с поврежденными ценами

```python
import re
data_t['Price'] = data_t['Price'].str.replace(',','')
data_t['Price'] = data_t['Price'].astype('float')
data_t.isnull().sum()
data_t = data_t.dropna()
```

установить размеры текста и изображения в seaborn для удобного просмотра

labelEncoder для каждого категориального столбца

```python
from sklearn.preprocessing import LabelEncoder

label_encoding_args = ['Neighborhood Group','Property Type', 'Room Type']
label_encoder = LabelEncoder()

for col in label_encoding_args:
  data_t[col] = label_encoder.fit_transform(data_t[col])+1
```

извлекать Features и цели в отдельных наборах данных

```python
target = data_t['Price']
features = data_t.drop({'Price'}, axis=1)
print(features.head())
```

разделить данные на обучение и тестирование

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_Y, test_Y = train_test_split(features, target, random_state = 47, test_size=1/3)
```

# обучение модели

тренируйте модель с помощью пула от Catboost, пул помогает определить категории.
(создание модели CatBoost и использование метода fit, как обычно, не дало хороших результатов, так как модель не знает, какие столбцы являются категориальными по умолчанию)

```python
from catboost import CatBoost, CatBoostRegressor, Pool
from sklearn.compose import ColumnTransformer



train_pool = Pool(train_X, 
                  train_Y,
                  cat_features=['Neighborhood Group', 'Property Type', 'Room Type']
                  )
test_pool = Pool(test_X, 
                 test_Y,
                 cat_features=['Neighborhood Group', 'Property Type', 'Room Type']
                )
model = CatBoostRegressor(learning_rate=0.2, 
                                     iterations =500, depth=14, 
                                     )

model.fit(train_pool, eval_set=test_pool, verbose=0, use_best_model=True, plot=True)
     

```

# метрики модели

метрики модели, в тестовых и обучающих данных

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error, accuracy_score

pred = model.predict(test_X)
pred_train = model.predict(train_X)


print("------------Catboost(test)---------------")
print("Mean squared error: %.2f" % mean_squared_error(test_Y, pred))
print("Mean absolute error: %.2f" % mean_absolute_error(test_Y, pred))
print("Coefficient of determination: %.2f" % r2_score(test_Y, pred))

print("------------Catboost(train)--------------")
print("Mean squared error: %.2f" % mean_squared_error(train_Y, pred_train))
print("Mean absolute error: %.2f" % mean_absolute_error(train_Y, pred_train))
print("Coefficient of determination: %.2f" % r2_score(train_Y, pred_train))
```

важность каждого столбца в модели

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(
    model, test_X, test_Y, n_repeats=10
)

sorted_importances_idx = result.importances_mean.argsort()
importances = pd.DataFrame(
    result.importances[sorted_importances_idx].T,
    columns=features.columns[sorted_importances_idx],
)
importances = importances[importances.columns[-10:]]
ax = importances.plot.box(vert=False, whis=10)
ax.set_title("Permutation Importances (test set)")
ax.axvline(x=0, color="k", linestyle="--")
ax.set_xlabel("Decrease in accuracy score")
ax.figure.tight_layout()

```
